```python
from transformers import AutoModelForCausalLM, set_seed, AutoTokenizer
import torch
```

```python
torch.set_default_tensor_type(torch.cuda.FloatTensor)
```

```python
model = AutoModelForCausalLM.from_pretrained("bigscience/bloom-1b7", use_cache=True)
tokenizer = AutoTokenizer.from_pretrained("bigscience/bloom-1b7")
```

```python
caption_generated= 'a little girl standing next to a wooden bench'
prompt =f'This is a  story  about  {caption_generated}  '
input_ids = tokenizer(prompt, return_tensors="pt").to(0)
```

```python
sample = model.generate(**input_ids, max_length=600, num_beams = 16, num_beam_groups = 4, top_k=1, temperature=8.3, repetition_penalty = 4.3, diversity_penalty = 0.9)
```

```python
print(tokenizer.decode(sample[0], truncate_before_pattern=[r"\n\n^#", "^'''", "\n\n\n"]))
```
